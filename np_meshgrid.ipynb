```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
X = np.array([1,2,3])
Y = np.array([1,2,3])
X, Y = np.meshgrid(X, Y)
```

```python
ax.plot(X, Y)
fig
```

```python
Z = X+Y
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X+Y
ax.scatter(X, Y, Z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.arange(0, 3, 1)
Y = np.arange(0, 3, 1)
Z = X+Y
ax.scatter(X, Y, Z, color='r', marker='p')
X, Y = np.meshgrid(X, Y)
Z = X+Y
ax.plot_surface(X, Y, Z, color='w')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
