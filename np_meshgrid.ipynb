```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
X = np.array([1,2,3])
Y = np.array([1,2,3])
X, Y = np.meshgrid(X, Y)
```
